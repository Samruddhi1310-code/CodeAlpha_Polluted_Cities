```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("indian_air_pollution_50_cities.csv")
```

```python
top10 = df.sort_values("aqi", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["city"], top10["aqi"])
plt.title("Top 10 Most Polluted Indian Cities (AQI)")
plt.xlabel("City")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.show()
```

```python
clean10 = df.sort_values("aqi").head(10)

plt.figure(figsize=(10,5))
plt.bar(clean10["city"], clean10["aqi"])
plt.title("Top 10 Cleanest Indian Cities (AQI)")
plt.xlabel("City")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.show()
```

```python
bins = [0,100,200,300,500]
labels = ["Good","Moderate","Poor","Very Poor"]
df["aqi_category"] = pd.cut(df["aqi"], bins=bins, labels=labels)

df["aqi_category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)
plt.title("Distribution of Indian Cities by AQI Level")
plt.ylabel("")
plt.show()
```
